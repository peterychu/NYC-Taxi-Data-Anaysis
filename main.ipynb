```python
import os 
import pandas as pd
```

```python
# Code to merge  and use data -- DELETE ONCE DB IS SET UP

jan = pd.read_parquet('data/parquet/yellow_tripdata_2023-01.parquet')
feb = pd.read_parquet('data/parquet/yellow_tripdata_2023-02.parquet')
mar = pd.read_parquet('data/parquet/yellow_tripdata_2023-03.parquet')
apr = pd.read_parquet('data/parquet/yellow_tripdata_2023-04.parquet')
may = pd.read_parquet('data/parquet/yellow_tripdata_2023-05.parquet')
jun = pd.read_parquet('data/parquet/yellow_tripdata_2023-06.parquet')
jul = pd.read_parquet('data/parquet/yellow_tripdata_2023-07.parquet')
aug = pd.read_parquet('data/parquet/yellow_tripdata_2023-08.parquet')
sep = pd.read_parquet('data/parquet/yellow_tripdata_2023-09.parquet')
oct = pd.read_parquet('data/parquet/yellow_tripdata_2023-10.parquet')
nov = pd.read_parquet('data/parquet/yellow_tripdata_2023-11.parquet')
dec = pd.read_parquet('data/parquet/yellow_tripdata_2023-12.parquet')

jan = jan.rename(columns={'airport_fee' : 'Airport_fee'})

months = [jan, feb, mar, apr, may, jun, jul, aug, sep, oct, nov ,dec]
df = pd.concat(months)
```

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

df['trip_id'] = df.index
```

```python
df.head()
```

```python
TaxiZoneLookup = pd.read_csv('data/taxi_zone_lookup.csv')

TaxiZoneLookup.head()
```

```python
datetime_dim = df[['tpep_pickup_datetime', 'tpep_dropoff_datetime']].copy()
datetime_dim = datetime_dim.reset_index(drop = True)

datetime_dim['PU_month'] = datetime_dim['tpep_pickup_datetime'].dt.month 
datetime_dim['PU_day'] = datetime_dim['tpep_pickup_datetime'].dt.day
datetime_dim['PU_hour'] = datetime_dim['tpep_pickup_datetime'].dt.hour
datetime_dim['PU_year'] = datetime_dim['tpep_pickup_datetime'].dt.year
datetime_dim['PU_dow'] = datetime_dim['tpep_pickup_datetime'].dt.weekday

datetime_dim['DO_month'] = datetime_dim['tpep_dropoff_datetime'].dt.month 
datetime_dim['DO_day'] = datetime_dim['tpep_dropoff_datetime'].dt.day
datetime_dim['DO_hour'] = datetime_dim['tpep_dropoff_datetime'].dt.hour
datetime_dim['DO_year'] = datetime_dim['tpep_dropoff_datetime'].dt.year
datetime_dim['DO_dow'] = datetime_dim['tpep_dropoff_datetime'].dt.weekday

datetime_dim['datetime_ID']  = datetime_dim.index

datetime_dim.head()
```

```python
payment_type_name_map = {1:'Credit Card', 2:'Cash', 3:'No charge', 
                         4:'Dispute', 5:'Unknown',  6:'Voidedtrip'}

payment_type_dim = df[['payment_type']]
payment_type_dim = payment_type_dim.reset_index(drop = True)
payment_type_dim['payment_type_ID'] = payment_type_dim.index
payment_type_dim['payment_type_name'] = payment_type_dim['payment_type'].map(payment_type_name_map)

payment_type_dim = payment_type_dim[['payment_type_ID','payment_type','payment_type_name']]

payment_type_dim.head()
```

```python
ratecode_name_map = {1:'Standard rate', 2:'JFK', 3:'Newark',
                     4:'Nassau or Westchester', 5:'Negotiated fare', 6:'Group ride'}

ratecode_dim = df[['RatecodeID']].copy()
ratecode_dim = ratecode_dim.reset_index(drop = True)
ratecode_dim['RateCode'] = ratecode_dim['RatecodeID']
ratecode_dim = ratecode_dim.drop(columns = ['RatecodeID'])
ratecode_dim['RateCodeID'] = ratecode_dim.index
ratecode_dim['RateCodeName'] = ratecode_dim['RateCode'].map(ratecode_name_map)

ratecode_dim = ratecode_dim[['RateCodeID','RateCode','RateCodeName']]

ratecode_dim.head()
```

```python
pickup_location_dim = df[['PULocationID']]
pickup_location_dim = pickup_location_dim.merge(TaxiZoneLookup, left_on =  'PULocationID', right_on = 'LocationID', how = 'left')
pickup_location_dim = pickup_location_dim.drop(columns = ['LocationID'])
pickup_location_dim = pickup_location_dim.rename(columns = {'Borough':'PU_Borough', 'Zone':'PU_Zone', 'service_zone':'PU_service_zone'})
pickup_location_dim.head()
```

```python
dropoff_location_dim = df[['DOLocationID']]
dropoff_location_dim = dropoff_location_dim.merge(TaxiZoneLookup, left_on =  'DOLocationID', right_on = 'LocationID', how = 'left')
dropoff_location_dim = dropoff_location_dim.drop(columns = ['LocationID'])
dropoff_location_dim = dropoff_location_dim.rename(columns = {'Borough':'DO_Borough', 'Zone':'DO_Zone', 'service_zone':'DO_service_zone'})
dropoff_location_dim.head()
```
